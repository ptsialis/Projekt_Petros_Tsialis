## Part 1 Extract Experiment 1

```python
for i in range(0,len(features_cleaned)):
    print(i)
    
        
    if isinstance(features_cleaned["chat_gpt_just_feature"][i], float):
        features_cleaned["just_feature_min_value"][i] = np.NaN
        
        features_cleaned["just_feature_max_value"][i] = np.NaN
        continue
        
        
    if "Infinity" in features_cleaned["chat_gpt_just_feature"][i] or "infinite" in features_cleaned["chat_gpt_just_feature"][i]:
        
        #m = re.search('(?<=Maximum:)(.*)', features_cleaned["chat_gpt_all_feature"][i])
        
        #x= m.groups()[0].split()
        
        
        features_cleaned["just_feature_max_value"][i]= np.NaN
        
        m = re.search('(?<=Minimum:)(.*)', features_cleaned["chat_gpt_just_feature"][2])
        features_cleaned["just_feature_min_value"][i] = m.groups()[0]
        
    
    elif "arbitrary" in features_cleaned["chat_gpt_just_feature"][i]:
        m = re.search('(?<=Minimum:)(.*)', features_cleaned["chat_gpt_just_feature"][2])
        features_cleaned["just_feature_max_value"][i] = m.groups()[0]
        
        features_cleaned["just_feature_max_value"][i] = np.NaN
        
    elif "unknown" in features_cleaned["chat_gpt_just_feature"][i] or "Unknown" in features_cleaned["chat_gpt_just_feature"][i]:
        features_cleaned["just_feature_min_value"][i] = np.NaN
        
        features_cleaned["just_feature_max_value"][i] = np.NaN
        
    elif "∞" in features_cleaned["chat_gpt_just_feature"][i] :
        features_cleaned["just_feature_min_value"][i] = np.NaN
        
        features_cleaned["just_feature_max_value"][i] = np.NaN  
        
    elif not any(char.isdigit() for char in features_cleaned["chat_gpt_just_feature"][i]):
        features_cleaned["just_feature_min_value"][i] = np.NaN
        
        features_cleaned["just_feature_max_value"][i] = np.NaN
    
        
    else:
        
        features_cleaned["chat_gpt_just_feature"][i] =features_cleaned["chat_gpt_just_feature"][i].replace(",", "")
        
        lis= re.findall(r"-?\d*\.\d+|-?\d+", features_cleaned["chat_gpt_just_feature"][i])
        
        if not lis or len(lis)> 2 or len(lis)< 2:
            
            features_cleaned["just_feature_min_value"][i] = np.NaN
            features_cleaned["just_feature_max_value"][i]= np.NaN
            
        else:
            print(lis)
            #print(features_cleaned["chat_gpt_just_feature"][i])
            features_cleaned["just_feature_min_value"][i] = lis[0]
            features_cleaned["just_feature_max_value"][i]= lis[1]
```
