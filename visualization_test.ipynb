## Calculation of JI

```python
def jaccard_t1(i):
    print(i)
    if np.isnan(df["just_feature_min_value"][i]):
       return np.NaN
    if np.isnan(df["just_feature_max_value"][i]):
       return np.NaN
    
    i1= interval([df["dataset_min_value"][i] , df["dataset_max_value"][i]])
    
    i2 = interval([df["just_feature_min_value"][i]  , df["just_feature_max_value"][i]])
    u = i1 | i2
    it =i1 & i2
    
    try:
        

        un1= u[0][1]-u[0][0]
        if len(it) == 0:
            
            np.NaN
        else:

            inter1= it[0][1]-it[0][0]

        if un1 == 0:
            return np.NaN
        else:
            
            iou=inter1/un1
            
        return iou
    
    except:
    
        return np.NaN
    
```

```python
for i in range(0,len(df)):
    
    jac = jaccard_t1(i)
    df["iou_just_feature"][i] = jac
    
```

```python
def jaccard_t2(i):
    
    if np.isnan(df["all_feature_min_value"][i]):
       return np.NaN
    if np.isnan(df["all_feature_max_value"][i]):
       return np.NaN
    
    i1= interval([df["dataset_min_value"][i] , df["dataset_max_value"][i]])
    
    i2 = interval([df["all_feature_min_value"][i]  , df["all_feature_max_value"][i]])
    
    u = i1 | i2
    it =i1 & i2
    
    try:
    
        
        un1= u[0][1]-u[0][0]
        if len(it) == 0:
            
            return np.NaN
        else:

            inter1= it[0][1]-it[0][0]
        
        
        if un1 == 0:
            return np.NaN
        else:
            
            
            iou=inter1/un1
        
        
        return iou
    
    except:
        
        return np.NaN
    
```

```python
for i in range(0,len(df)):
    
    jac = jaccard_t2(i)
    df["iou_all_feature"][i] = jac
    
```
